### 제목과 내용에 따른 신뢰도 등급 분류 학습
- 사용데이터 : trade_classification_data.csv (ai 생성)

```python
import pandas as pd

post_df = pd.read_csv('./datasets/trade_classification_data.csv')
post_df
```

```python
post_df.info()
```

```python
post_df.isna().sum()
```

```python
post_df = post_df[~post_df.title.isna()]
post_df = post_df[~post_df.content.isna()]
post_df
```

```python
post_df.duplicated().sum()
```

```python
post_df.drop_duplicates(inplace=True, ignore_index=True)
post_df
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
targets = encoder.fit_transform(post_df.target)

post_df.target = targets
post_df
```

```python
post_df.target.value_counts()
```

```python
targets_2 = post_df[post_df.target == 2].sample(2144, random_state=124)
targets_1 = post_df[post_df.target == 1].sample(2144, random_state=124)
targets_0 = post_df[post_df.target == 0]

post_df = pd.concat([targets_0, targets_1])
post_df = pd.concat([post_df, targets_2])
post_df.reset_index(drop=True, inplace=True)
post_df
```

```python
post_df.target.value_counts()
```

```python
pre_p_df = post_df.copy()
pre_p_df['text'] = pre_p_df.title + " " + pre_p_df.content
pre_p_df = pre_p_df.drop(['title', 'content'], axis=1)
pre_p_df
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from kiwipiepy import Kiwi

kiwi = Kiwi()
josa_tags = ["JKS","JKC","JKO","JKG","JKB","JKV","JKQ","JX","JC",
            "EC","EF","EP","ETN","ETM"]

def removes_josa(sentence):
    result = []
    for token in kiwi.tokenize(sentence):
        if token.tag in josa_tags:
            continue

        result.append(token.lemma)
    return " ".join(result)

pre_p_df.text = pre_p_df.text.apply(removes_josa)
pre_p_df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = \
train_test_split(pre_p_df.text, pre_p_df.target, stratify=pre_p_df.target, test_size=0.2, random_state=124)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

m_nb_pipe = Pipeline(
    [
        ('count_vectorizer', CountVectorizer()), 
        ('multinomial_NB', MultinomialNB(alpha=0.2))
    ]
)

m_nb_pipe.fit(X_train.values, y_train)
```

```python
m_nb_pipe.score(X_test.values, y_test)
```
