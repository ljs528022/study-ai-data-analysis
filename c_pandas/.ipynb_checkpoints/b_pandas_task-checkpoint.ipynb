```python
import pandas as pd
import numpy as np
from scipy.stats import iqr
# 가설: 모든 요소 중 어떤 수치가 가장 행복지수에 영향이 클까?

happiness_df = pd.read_csv('./datasets/happiness_report_2023.csv')

print(happiness_df.shape)
print(happiness_df.info())
print(happiness_df.describe())
```

```python
# 결측치 제거
happiness_df = happiness_df[~happiness_df.healthy_life_expectancy.isna()].reset_index(drop=True)
happiness_df
```

```python
# 이상치 제거
# 숫자값인 모든 값을 list로 모아놓기
numeric_cols = [
    'happiness_score', 'gdp_per_capita', 'social_support',
    'healthy_life_expectancy', 'freedom_to_make_life_choices',
    'generosity', 'perceptions_of_corruption'
]

# 이상치 계산
happiness_Q1 = np.percentile(happiness_df[numeric_cols], 25, axis=0)
happiness_Q3 = np.percentile(happiness_df[numeric_cols], 75, axis=0)
iqr_value = iqr(happiness_df[numeric_cols], axis=0)
# print(happiness_Q1)
# print(happiness_Q3)
# print(iqr_value)

lower_bound = happiness_Q1 - 1.5 * iqr_value
upper_bound = happiness_Q3 + 1.5 * iqr_value
# print(lower_bound)
# print(upper_bound)

condition = (
    (happiness_df[numeric_cols] >= lower_bound) &
    (happiness_df[numeric_cols] <= upper_bound)
).all(axis=1)

# 이상치 제거
happiness_df = happiness_df[condition].reset_index(drop=True)
happiness_df
```

```python
# 그룹별 평균 비교

# gdp_per_capita를 높음/낮음으로 나눠서 행복점수 평균 비교
feature_cols = [
    'gdp_per_capita', 'social_support', 'healthy_life_expectancy',
    'freedom_to_make_life_choices', 'generosity', 'perceptions_of_corruption'
]

for col in feature_cols:
    # 중앙값 기준 그룹 나누기
    median_val = happiness_df[col].median()
    happiness_df['group'] = '낮음'
    happiness_df.loc[happiness_df[col] >= median_val, 'group'] = '높음'

    # 그룹별 평균 출력
    group_mean = happiness_df.groupby('group')['happiness_score'].mean()
    print(f"[{col}]")
    print(group_mean)
    print()
```

```python
# 시각화
fig, axes = plt.subplots(2, 3, figsize=(12, 9))
axes = axes.flatten()

# axes를 1차원 배열로 만들어, for문으로 삽입
for i in range(len(feature_cols)):
    col = feature_cols[i]
    happiness_df.plot(title=col, x=col, y='happiness_score', fontsize='7', kind='scatter', grid=True, ax=axes[i])

# 시각화
plt.show()
```

#### 행복에 가장 큰 영향을 미치는 요소 분석 보고서
- gdp_per_capita (1인당 GDP)의 경우, 산점도에서 뚜렷한 우상향 추세를 보인다.
높음 그룹의 평균 행복점수는 6.11, 낮음 그룹은 4.84로 약 1.27점의 차이를 보였으며,
이는 1인당 GDP가 높을수록 행복 점수도 높아지는 경향이 있음을 의미한다.

- social_support (사회적 지지)의 경우, 6개 요소 중 가장 뚜렷한 우상향 추세를 보인다.
높음 그룹의 평균 행복점수는 6.17, 낮음 그룹은 4.79로 약 1.38점의 차이를 보였으며,
이는 6개 요소 중 가장 큰 차이로, 사회적 지지가 행복에 가장 큰 영향을 미친다는 것을 시사한다.

- healthy_life_expectancy (건강 기대수명)의 경우, 우상향 추세가 명확하게 나타난다.
높음 그룹의 평균 행복점수는 6.08, 낮음 그룹은 4.85로 약 1.23점의 차이를 보였으며,
건강한 삶이 행복과 밀접한 관계가 있음을 시사한다.

- freedom_to_make_life_choices (삶의 자유)의 경우, 약한 우상향 추세를 보인다.
높음 그룹의 평균 행복점수는 6.00, 낮음 그룹은 4.96으로 약 1.04점의 차이를 보였으며,
앞선 세 요소에 비해 그 차이가 작아 행복과의 관계가 상대적으로 약함을 나타낸다.

- generosity (관대함)의 경우, 산점도에서 점들이 흩어져 있으며 뚜렷한 추세를 보이지 않는다.
오히려 높음 그룹의 평균 행복점수는 5.28, 낮음 그룹은 5.68로 낮음 그룹이 더 높게 나타났으며, 이는 관대함이 행복 점수와 뚜렷한 관계가 없음을 의미한다.

- perceptions_of_corruption (부패 인식)의 경우, 산점도에서 점들이 전반적으로 흩어져 있다.
높음 그룹의 평균 행복점수는 5.59, 낮음 그룹은 5.36으로 약 0.23점의 차이에 불과하며,
이는 부패 인식이 행복에 미치는 영향이 미미함을 시사한다.
