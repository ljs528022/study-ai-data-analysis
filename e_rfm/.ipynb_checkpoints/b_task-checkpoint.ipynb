### RFM 분석 실습
- Recency: 얼마나 최근에 구매했는가
- Frequency: 얼마나 자주 구매했는가
- Monetary: 얼마나 많은 금액을 지출했는가

### 고객 분석
`customer_order_data.csv`
- CustomerID: 고객 번호
- PurchaseDate: 구매 날짜
- TransactionAmount: 결제 금액 
- ProductInformation: 상품 정보
- OrderID: 주문 번호
- Location: 주문 지역

#### 중간 정리
- 비교를 위해 사용될 수 있는 feature : TransactionAmount(결제 금액), ProductInformation(상품 정보), Location(주문 지역)
- '범주'로 사용될 수 있는 feature : PurchaseDate(구매 날짜), OrderID(주문 번호)

### RFM 값 설정
- Recency(얼마나 최근에?) : PurchaseDate의 max 값을 조회해보고, max 값을 기준으로 각 날짜에 따른 '일차'를 계산
- Frequency(얼마나 자주?) : 회원별로 중복치를 조회해서 나온 수.
- Monetary(얼마나 많이?) : 중복된 회원을 조회해보고, 해당 회원의 모든 거래에 대한 TransactionAmount을 합친 값.

```python
# RFM의 'Recency' 값 계산해보기

# min, max 값 조회해보기
print(f'min : {co_df.PurchaseDate.min()}')
print(f'max : {co_df.PurchaseDate.max()}')

# 조회 결과
# 해당 데이터는 4월 ~ 6월 사이 약 2개월 간의 거래 정보에 대한 데이터임을 확인
```

```python
# 계산 전에 회원 별 거래 개수 조회
co_df.CustomerID.value_counts()

# 앞으로의 계산에서 회원별로 그룹화가 필요해보임
```

```python
# 가장 최근 값인 2023-06-10을 기준으로 일차 계산하기

# 저장할 co_rfm_df 생성
co_rfm_df = pd.DataFrame()

# 최신 일자
reference_date = pd.to_datetime(co_df['PurchaseDate']).max()
# 최신 일자를 기준으로 각 회원별 거래일자를 뺸 일차를 계산 후, 저장
co_rfm_df['Recency'] = co_df.groupby('CustomerID')['PurchaseDate'].apply(lambda x: (reference_date - pd.to_datetime(x).max()).days)
co_rfm_df.reset_index(inplace=True)
co_rfm_df
```

```python
# RFM의 'Frequency' 값 계산해보기

# CustomerID 기준으로 회원의 이용빈도를 조회
print(co_df.CustomerID.value_counts().sort_values(ascending=False))

co_rfm_df['Frequency'] = co_df.groupby('CustomerID')['OrderID'].count().reset_index(drop=True)
co_rfm_df
```

```python
# RFM의 'Monetary' 값 계산해보기

# 회원별로 그룹화하여 모든 거래 금액을 합산
co_rfm_df['Monetary'] = co_df.groupby('CustomerID')['TransactionAmount'].sum().reset_index(drop=True)
co_rfm_df
```

```python
# 구한 RFM 값을 정규화
from sklearn.preprocessing import MinMaxScaler

normalization = MinMaxScaler()
rfm_normalization = normalization.fit_transform(co_rfm_df[['Recency','Frequency', 'Monetary']])
rfm_normalization
```

```python
# 구한 정규화 값을 다시 DataFrame으로 변경
co_rfm_norm_df = pd.DataFrame(rfm_normalization, columns=['Recency','Frequency', 'Monetary'])
co_rfm_norm_df
```

```python
# Recency 값 재조정
co_rfm_norm_df.Recency = 1 - co_rfm_norm_df.Recency
co_rfm_norm_df
```

```python
# RFM 값을 통한 Total 점수 계산하기
co_rfm_df['Total'] = co_rfm_norm_df.Recency + co_rfm_norm_df.Frequency + co_rfm_norm_df.Monetary
co_rfm_df
```

```python
# co_result_df 에 분석에 사용할 feature과 Total을 merge 한다.
# 사용하는 feature : 
co_result_df = co_df.merge(co_rfm_df[['CustomerID','Total']], on='CustomerID')
co_result_df
```

```python
import numpy as np

# Total 값을 3분위로 분배한다.
g2, g1 = np.percentile(co_rfm_df.Total, [40, 80])
print(g2, g1)
```

```python
def get_grade(x):
    if x <= g2:
        return 3
    if x <= g1:
        return 2
    else:
        return 1
```

```python
# Total 값에 따른 회원별 등급 부여
co_result_df['Grade'] = co_result_df.Total.apply(get_grade).replace([1, 2, 3], ['Gold','Silver','Bronze'])
co_result_df
```

```python
# 시각화
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'

# 등급 순으로 뿌리기 위한 list
order_columns = ['Bronze', 'Silver', 'Gold']

fig, axes = plt.subplots(2, 2, figsize=(12, 6))

# 회원을 등급 순으로 count
sns.countplot(ax=axes[0][0], data=co_result_df, x='Grade', palette='muted', order=order_columns)
sns.countplot(ax=axes[0][1], data=co_result_df, x='Grade', hue='Location', palette='muted', order=order_columns)
plt.show()
```

```python
# 어떤 지역에 어떤 등급의 회원이 있을까?
sns.countplot(data=co_result_df, x='Grade', hue='Location', palette='muted', order=order_columns)
plt.title('등급별 고객 분포')
plt.show()
```
