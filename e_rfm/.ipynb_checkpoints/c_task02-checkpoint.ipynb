### RFM 분석 실습
- Recency: 얼마나 최근에 구매했는가
- Frequency: 얼마나 자주 구매했는가
- Monetary: 얼마나 많은 금액을 지출했는가
  

### 고객 분석
`customer_invoice_data.csv`
- InvoiceNo: 거래 번호 (유심히 살펴볼 것!)
- StockCode: 상품 코드
- Description: 상품명 
- InvoiceDate: 거래 날짜
- UnitPrice: 상품 단가
- CustomerID: 고객 번호
- Country: 국가명

```python
import pandas as pd

ci_df = pd.read_csv('./datasets/customer_invoice_data.csv')
ci_df
```

```python
ci_df.describe().T

# 음수값이 있다...?
```

```python
display(ci_df[ci_df.CustomerID.isna()])
# 회원이 누군지 판별할 수 없으면 데이터를 활용할 수 없다. -> 삭제
ci_df = ci_df[~ci_df.CustomerID.isna()].reset_index(drop=True)
ci_df
```

```python
# 데이터 중, InvoiceNo의 값 맨 앞에 'C'가 붙어있는 값이 있는 것을 확인했다.
# 무슨 데이터들인지 확인하자.

ci_df[ci_df.InvoiceNo.str.contains('C')]
# 확인결과, C가 붙은 것들은 '취소된 거래' 인 것으로 확인되었다.
```

```python
# 편하게 columns 부터 바꾸자.
ci_df.columns = ['InvoiceNo', 'StockCode', 'Description', 'Quantity', 'Date', 'Price', 'CustomerID', 'Country']
ci_df

# Date의 값을 date 타입으로 변경
ci_df.Date = pd.to_datetime(ci_df.Date)
ci_df

# Monetary 값을 위한 TotalPrice 미리 생성
ci_df['TotalPrice'] = ci_df.Price * ci_df.Quantity
ci_df

# RFM 값을 구하기 전에 거래 건수 중에 '취소건'을 먼저 삭제한다.
ci_df = ci_df[~ci_df.InvoiceNo.str.contains('C')]
ci_df
```

```python
# Recency 구하기
# 가장 최근 일자에서 각 일자를 빼자
target = ci_df.Date.max()
ci_df['Recency'] = (target - pd.to_datetime(ci_df.Date)).dt.days
ci_df
```

```python
# 각 회원의 거래 수 조회
# 데이터 구조 상, 한번에 거래에 여러개의 상품을 주문이 있고, 데이터는 각 상품마다 한 행을 차지한다.
# 그럼 회원의 거래빈도를 구하려면 InvoiceNo 조회해보자.
print(ci_df.InvoiceNo.value_counts())

# 한번의 거래에서 최대 542개의 상품을 구매하는 회원이 존재했다.
# 해당 데이터의 거래는 총 22190건이다.
# 여기서, 취소된 건수는 3654건이였다. -> 전처리 과정애서 삭제되었다.
```

```python
# rfm_df 생성
rfm_r = ci_df.groupby('CustomerID')['Recency'].min().reset_index()
rfm_r

# Frequency 구하기
# 회원 별로 거래건수를 조회해야한다.
rfm_f = ci_df.groupby('CustomerID')['InvoiceNo'].apply(lambda x: len(x.unique())).reset_index()
rfm_f.rename(columns={'InvoiceNo': 'Frequency'}, inplace=True)
rfm_f

# Monetary 구하기
# 회원별로 미리 구해놓은 TotalPrice 값을 합산하면 된다.
rfm_m = ci_df.groupby('CustomerID')['TotalPrice'].sum().reset_index()
rfm_m.rename(columns={'TotalPrice': 'Monetary'}, inplace=True)
rfm_m

rfm = pd.merge(rfm_r, rfm_f, on='CustomerID')
rfm = rfm.merge(rfm_m, on='CustomerID')
rfm
```

```python
# 졍규화
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

rfm[['Recency', 'Frequency', 'Monetary']] = scaler.fit_transform(rfm[['Recency', 'Frequency', 'Monetary']])
rfm
```

```python
rfm.Recency = 1 - rfm.Recency
rfm
```

```python
# TotalScore 계산
rfm['TotalScore'] = rfm.Recency + rfm.Frequency + rfm.Monetary
rfm
```

```python
ci_df.drop(labels='Recency', axis=1, inplace=True)
```

```python
# 원본 데이터와 merge
ci_rfm_df = ci_df.merge(rfm, on='CustomerID')
ci_rfm_df
```

```python
import numpy as np

g4, g3, g2, g1 = np.percentile(ci_rfm_df.TotalScore, [25, 35, 55, 75])
print(g4, g3, g2, g1)      
# 0.05, 0.06, 0.05 차이
```

```python
def get_grade(x):
    if x <= g4:
        return 5
    if x <= g3:
        return 4
    if x <= g2:
        return 3
    if x <= g1:
        return 2
    else:
        return 1
```

```python
ci_rfm_df['Grade'] = ci_rfm_df['TotalScore'].apply(get_grade).replace([1, 2, 3, 4, 5], ['VIP', 'Diamond', 'Gold', 'Silver', 'Bronze'])
ci_rfm_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'

order = ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP']

print(ci_rfm_df.groupby('Grade')['CustomerID'].nunique())

customer_df = ci_rfm_df.drop_duplicates(subset='CustomerID').reset_index()

# 등급별 회원별 분포
sns.countplot(data=customer_df, x='Grade', palette='muted', order=order)
plt.title('Grade')
plt.show()
```

```python
# 등급별 나라 분포
top_countries = customer_df['Country'].value_counts().head(10).index
top_countries

country_df = customer_df[customer_df.Country.isin(top_countries)]
sns.countplot(data=country_df, x='Grade', palette='muted', order=order)
```
